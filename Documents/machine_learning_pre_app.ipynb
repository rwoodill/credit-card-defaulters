```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import pandas as pd 
```

```python
file_path_ad_clean = "../Raw_data/pre_application_data_cleaned.csv"
application_df = pd.read_csv(file_path_ad_clean)
application_df.head()
```

```python
# convert text to numbers for model
application_df = pd.get_dummies(application_df)
```

```python
application_df.head(10)
```

```python
#seperate feature and target
features = application_df.drop('NAME_CONTRACT_STATUS',axis=1).values
target = application_df['NAME_CONTRACT_STATUS'].values
```

```python
X = features
y = target
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 5)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features = len(X_train_scaled[0])
nodes_layer1 = 80
nodes_layer2 = 30
nodes_layer3 = 1

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units = nodes_layer1, input_dim = num_input_features, activation ="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units = nodes_layer2, activation ="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units = nodes_layer3, activation ="sigmoid"))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=20)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
