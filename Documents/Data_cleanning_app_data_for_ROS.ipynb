Data Cleaning

```python
file_path_ad = "../Raw_data/application_data.csv"
#file_path_pad = "../Raw_data/previous_application.csv"
```

```python
import pandas as pd 
from pprint import pprint
df = pd.read_csv(file_path_ad)
df.head(20)
```

```python
# Count null values for each column
null_count_list = []
for column in df.columns:
    null_count = df[column].isnull().sum()
    if null_count > 0:
        null_count_list.append(f"{column}: {null_count} null values")
        print(f"{column}: {null_count} null values")
null_count_list_df = pd.DataFrame(null_count_list)
file_path_null = "../Raw_data/application_data_null_list_app.csv"
null_count_list_df.to_csv(file_path_null,index=False)
```

```python
# Narrow down the NA search
for column in df.columns:
    null_count = df[column].isnull().sum()
    if 1000> null_count > 0:
        print(f"{column}: {null_count} null values")
```

```python
# Narrow down the NA search
null_count_list_200k = []
for column in df.columns:
    null_count = df[column].isnull().sum()
    if null_count > 200000:
        print(f"{column}: {null_count} null values")
        null_count_list_200k.append(column)
print(null_count_list_200k)
```

```python
columns_to_check = ['AMT_ANNUITY', 'AMT_GOODS_PRICE','NAME_TYPE_SUITE','CNT_FAM_MEMBERS','EXT_SOURCE_2','DAYS_LAST_PHONE_CHANGE']
df_cleaned = df.dropna(subset=columns_to_check)
```

```python
# drop customer ID
df_cleaned = df_cleaned.drop('SK_ID_CURR',axis=1)
# drop columns with over 200k NA
for item in null_count_list_200k:
    df_cleaned = df_cleaned.drop(item,axis=1)
```

```python
# Count null values for each column, check if they exist in cleand data
for column in df_cleaned.columns:
    null_count = df_cleaned[column].isnull().sum()
    if 1000 > null_count > 0:
        print(f"{column}: {null_count} null values")
for column in df_cleaned.columns:
    null_count = df[column].isnull().sum()
    if null_count > 200000:
        print(f"{column}: {null_count} null values")
```
